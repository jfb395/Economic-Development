```python
import math
# defining the sigmoidal and linear functions needed to simulate the trap
def sigmoid(x):
    a = []
    for item in x:
        a.append(1/(1+math.exp(-item)))
    return a
def linear(x):
    b = []
    for item in x:
        b.append(.1*(item+5))
    return b

import matplotlib.pyplot as plt
import numpy as np

#graphing the lines 
x = np.arange(-10., 10., 0.2)
sig = sigmoid(x)
lin = linear(x)
plt.plot(x,sig)
plt.plot(x,lin,label='H(t)=H(t+1)')
plt.legend()
plt.ylim(ymax=1.1,ymin=-.1)
plt.xlim(xmax=6,xmin=-6.0)

#Labeling the axes 
labels=['H(undeveloped)','','','H*','','H(developed)']
labels2=['','','','','','','','']
plt.xticks(np.arange(-6,6,2),labels)
plt.yticks(np.arange(0,1.1,.1),labels2,size=8)
plt.xlabel('Human Capital Stock(t)')
plt.ylabel('Human Capital Stock(t+1)')
plt.suptitle('Demographic Poverty Trap: Niger \n Human Capital Stock', fontsize=12, fontweight='bold')

#Creating the Horizontal and Veritcal Lines 
plt.axhline(y=0.5, color='r', linestyle=':',alpha=.5)
plt.axvline(x=0.0, color='r', linestyle=':',alpha=.5)


#Annotating Niger
plt.annotate("Niger",xy=(-4,-.07),color='m')
bbox_props_niger = dict(boxstyle="round,pad=0.3", fc="magenta", ec="k", lw=.5,alpha=.5)
t_niger = plt.text(-1,0, ".51% w/ teritary education", ha="center", va="center",
            size=8,
            bbox=bbox_props_niger)



#Creating the Dots on the Graph
plt.plot(-4.95,.01,'o',color='r',markersize=10)
plt.plot(4.92,.999,'o',color='g',markersize=10)
plt.plot(0,.5,'o',color='b',markersize=10)
plt.plot(-4,.02,'d',color='m',markersize=12)

#Creating the boxes for the steady-state labels
bbox_props_undev = dict(boxstyle="round,pad=0.3", fc="red", ec="k", lw=.5,alpha=.5)
t_undev = plt.text(-4.6,.09, "H(Undeveloped)", ha="center", va="center",
            size=8,
            bbox=bbox_props_undev)

bbox_props_dev = dict(boxstyle="round,pad=0.3", fc="green", ec="k", lw=.5,alpha=.5)
t_dev = plt.text(3.5,1.04, "H(Developed)", ha="center", va="center",
            size=8,
            bbox=bbox_props_dev)

bbox_props_inflect = dict(boxstyle="round,pad=0.3", fc="blue", ec="k", lw=.5,alpha=.3)
t_inflect = plt.text(1.7,.40, "20% w/ \n tertiary education", ha="center", va="center",
            size=8,
            bbox=bbox_props_inflect)

bbox_props_inf = dict(boxstyle="round,pad=0.3", fc="blue", ec="k", lw=.5,alpha=.3)
t_inf = plt.text(-1.5,.55, "(Inflection Point)", 
                  ha="center", va="center",
            size=8,
            bbox=bbox_props_inf)

#plt.annotate("(Inflection Point)",xy=(-3.3,.55),color='b')

# Arrows showing dynamics
d = np.arange(-1.6, -.1, 0.2)
plt.plot(d,sigmoid(d),'r>',markersize=6)

e = np.arange(.2, 1.2, 0.2)
plt.plot(e,sigmoid(e),'g<',markersize=6)


plt.show()
```
